Il calcolo dell'intensità integrale media è data dalla somma dei valori dei pixel che mi costituiscono la sferetta, diviso per il numero totale di pixel. Per capire se un pixel sia o no appartenente alla sferetta metto un if che mi controlli che non sia 0 il suo valore (non blu, cioè circa 20, perchè ho sottratto il parametro costante mode alla mia immagine)

```python
# CALCOLO DELL'INTENSITA'INTEGRALE MEDIA

def intensity(x, y, imm):
    z = []    
    delta = 5
    for xi, yi in zip(x, y):
        val=0
        count=0
        for i in range(-delta,delta):
            for j in range(-delta,delta):
                if imm[yi+i,xi+j]>0:
                    count=count+1
                    val = val + imm[yi+i,xi+j]
        z.append(val/count)  
    z = np.array(z)
    return z
```

La ricerca dei massimi avviene tramite l'applicazione di 2 filtri di max e min che agiscono su dei quadrati di lato vicinanza_size (stimabile con il raggio della sferetta). I due filtri vengono sottratti e se la loro differenza è maggiore di un valore fissato (soglia) allora viene riconosciuto come max 

```python
# RICERCA DEI MASSIMI 

def find_max(imm):
    vicinanza_size = 10
    
    imm_max = filters.maximum_filter(imm, vicinanza_size)
    maxima = (imm == imm_max)
    imm_min = filters.minimum_filter(imm, vicinanza_size)
    
    soglia = 40
    
    diff = ((imm_max - imm_min) > soglia)
    maxima[diff == 0] = 0
    
    labeled, num_objects = ndimage.label(maxima)
    
    slices = ndimage.find_objects(labeled)
    
    margine = 50
    
    x, y = [], []
    for dy,dx in slices:
        x_center = (dx.start + dx.stop - 1)//2
        y_center = (dy.start + dy.stop - 1)//2 
        
        if x_center<margine or x_center>imm.shape[1]-margine: #ESCLUDO I PUNTI SUL BORDO
            continue                                         
            
        if y_center<margine or y_center>imm.shape[0]-margine:
            continue
        
        x.append(x_center)
        y.append(y_center)
    
    x = np.array(x)
    y = np.array(y)
    
    z = intensity(x, y, imm)
     
    return x, y, z
```

E' la forma ipotizzata della superficie che fitta al meglio i punti di massimo. Si ipotizza sia una gaussiana ma in realtà troviamo un esponenziale (e è circa 3,2)

```python
# RICERCA DELLA FUNZIONE CHE MI FITTA I MASSIMI

def gauss(x, y, dsx, dsy, maxint, corr, cx, cy, bg, e):
        if dsx<0:
            return np.nan
        if dsy<0:
            return np.nan
        if maxint<0:
            return np.nan
        if cx<0:
            return np.nan
        if cy<0:
            return np.nan
        if bg<0: 
            return np.nan
        if ((corr<-1) or (corr>1)):
            return np.nan
        if e<0:
            return np.nan
            
        varx =((x-cx)/dsx)**2
        vary = ((y-cy)/dsy)**2
        varxy = (corr*(x-cx)*(y-cy))/(dsy*dsx)
        d = sqrt(varx+vary+varxy)
        
        return ne.evaluate('bg+maxint*exp(-0.5*d**e)') # PARALLELIZZATO
```

```python
def analizza(imm):

    x, y, z = find_max(imm)
    
    xy = array(zip(x, y))
    
    def gauss_for_curve(xy, *args):
        x, y = xy.T
        return gauss(x, y, *args)
    
    parametri_0 = (800, 600, np.max(imm), 0, 800, 600, 400, 2)
    parametri, covar = curve_fit(gauss_for_curve, xy, z, parametri_0)
    imm_normalized = correzione(imm, parametri)
        #FIT RIFATTO SENZA I PUNTI PIU INTENSI
    quantile = mquantiles(imm_normalized[y, x], 0.95)
    
    newx, newy = [], []
    
    for (valore, xtemp, ytemp) in zip(imm_normalized[y, x], x, y):
        if valore < quantile:
            newy.append(ytemp) 
            newx.append(xtemp)
    
    newx = asarray(newx) 
    newy = asarray(newy)

    newz = intensity(newx, newy, imm)
    
    newxy = array(zip(newx, newy))
    
    newparametri, newcovar = curve_fit(gauss_for_curve, newxy, newz, parametri)
    print newparametri
    return newparametri
```

```python
def correzione(imm, par):

    xx, yy = np.meshgrid(np.linspace(0, 1600, 1600), np.linspace(0, 1200, 1200))
    
    zz = gauss(xx, yy, *par)
    
    zz_normalized = zz / np.max(zz)
    imm_normalized = imm/zz_normalized
    return imm_normalized
```

```python
# RICERCA DEI MASSIMI(nuclei) - ha i parametri vicinanza_size e soglia diversi

def find_max_cell(imm):
    vicinanza_size = 100
    
    imm_max = filters.maximum_filter(imm, vicinanza_size)
    maxima = (imm == imm_max)
    imm_min = filters.minimum_filter(imm, vicinanza_size)
    
    soglia = 300
    
    diff = ((imm_max - imm_min) > soglia)
    maxima[diff == 0] = 0
    
    labeled, num_objects = ndimage.label(maxima)
    
    slices = ndimage.find_objects(labeled)
    
    margine = 50
    
    x, y = [], []
    for dy,dx in slices:
        x_center = (dx.start + dx.stop - 1)//2
        y_center = (dy.start + dy.stop - 1)//2 
        
        if x_center<margine or x_center>imm.shape[1]-margine: #ESCLUDO I PUNTI SUL BORDO
            continue                                         
            
        if y_center<margine or y_center>imm.shape[0]-margine:
            continue
        
        x.append(x_center)
        y.append(y_center)
    
    x = np.array(x)
    y = np.array(y)
    
    z = intensity(x, y, imm)
     
    return x, y, z
```
